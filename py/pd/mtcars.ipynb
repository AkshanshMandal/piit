```python
# libraries
import numpy as np
import pandas as pd
from pydataset import data
```

```python
mtcars = data('mtcars')
mtcars.head()
```

```python
mtcars.dtypes
```

```python
# Convert selected columns to category
mtcars = data('mtcars')
mtcars = mtcars.astype({ 'cyl':'category', 'vs': 'category',  'carb': 'category',  'am':   'category',  'gear':  'category' } )
```

```python
mtcars.dtypes
```

```python
#changing am to auto and manual
mtcars['am']=mtcars['am'].map({0:'auto',1:'manual'})
mtcars.head()
```

```python
#changing vs to vshaped and stright
mtcars['vs'] = mtcars['vs'].map({0: 'Vshaped', 1: 'Straight'})
mtcars.head()
```

```python
mtcars.index
```

```python
#adding index as column with column name as car_name
mtcars['car_name']=mtcars.index
```

```python
mtcars.head()
```

```python
print(len(mtcars[mtcars['mpg']>25]))
```

```python
#printing mtcars with mpg greater than 25
#1)
mtcars[mtcars['mpg']>25]
```

```python
#2)
mtcars.query("mpg > 25")
```

```python
#showing columns mpg and hp using filter
mtcars.filter(items=['mpg','hp']).head()
```

```python
#listing all the cars which is auto
mtcars[mtcars['am']=='auto'].head()
```

```python
# Use & for “AND” between them. cars = auto, cyl <> 5, mpg>22 ; all 3 conditions
mtcars[(mtcars['am']=='auto') & (mtcars['cyl']!='5') & (mtcars['mpg']>22)]
```

```python
# cars with mpg between 23 and 26
mtcars[mtcars['mpg'].between(left=23, right=26.1,inclusive='left')]
```

```python
mtcars[mtcars['mpg'].between(20,30, inclusive = 'neither')]
```

```python
mtcars[mtcars['mpg'].between(20,30) & (mtcars['am']=='manual')]
```

```python
#mtcars['gear'].isin([4,5])
# list cars whose gear is in values 4,5
mtcars[mtcars['gear'].isin([4,5])]
```

```python
# all auto or gear = 4,5
mtcars[ (mtcars['gear'].isin([4, 5])) | (mtcars['am'] =='auto')]
```

```python
# cars other gears values in 4,5 : ~ opposite
mtcars[~mtcars['gear'].isin([4, 5])]
```

```python
pd.pivot_table( observed=False , data= mtcars,  index='gear',  columns='am',  values='mpg', \
                aggfunc=['mean', 'max', 'min' ,'count','sum'] ).reset_index()
```
