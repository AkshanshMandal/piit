# Hypothesis T-Test

```python
import numpy as np
from scipy import stats
```

```python
# Generate a sample of 20 widget weights
np.random.seed(42)  # for reproducibility
sample_weights = np.random.normal(loc=51.5, scale=2, size=20)
print("Sample weights:", np.round(sample_weights, 2))
```

```python
# Define the hypothesized population mean
pop_mean = 50

# Perform the one-sample t-test
t_statistic, p_value = stats.ttest_1samp(a=sample_weights, popmean=pop_mean)

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
```

- Ho : mu = 50 (if pvalue > 0.05) : failed to reject Ho
- Ha : mu != 50 (if pvalue < 0.05); reject Ho infavour of Ha

```python
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis", p_value)
    print("The sample mean is significantly different from the hypothesized population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis.", p_value)
    print("There is no significant difference between the sample mean and the hypothesized population mean.")
```
